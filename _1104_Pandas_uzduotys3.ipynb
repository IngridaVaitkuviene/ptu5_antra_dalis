# 9 Parašykite funkciją, kuri mėnesio pavadinimą verstų skaičiumi. Naudodami .apply() lentelėje pakeiskite mėnesių pavadinimus skaičiais. 
*nusikopijuokite žodyną, kad nereikėtų rašyti ranka :)*

```python
menesiai = {1: 'January',
  2: 'February',
  3: 'March',
  4: 'April',
  5: 'May',
  6: 'June',
  7: 'July',
  8: 'August',
  9: 'September',
  10: 'October',
  11: 'November',
  12: 'December'}
```

```python
def menesiai_skaiciais(menesis):
    menesiai = {1: 'January',
  2: 'February',
  3: 'March',
  4: 'April',
  5: 'May',
  6: 'June',
  7: 'July',
  8: 'August',
  9: 'September',
  10: 'October',
  11: 'November',
  12: 'December'}
    for skaicius, men in menesiai.items():
        if men == menesis:
            return skaicius
```

```python
print(menesiai_skaiciais('March'))
```

```python
fires['month'] = fires['month'].apply(menesiai_skaiciais)
```

* **Pakeiskite indeksą į stulpelį 'Miestas'**
* **Sutvarkykite duomenis taip, kad gyventojų sk. reprezentuotų langelis su 'int' reikšme: -
Ten kur trūksta duomenų - nulis.

```python
miestai_url = pd.read_html('https://lt.wikipedia.org/wiki/S%C4%85ra%C5%A1as:Lietuvos_miestai_pagal_gyventojus')
miestai_url
```

```python
len(miestai_url)
```

```python
miestai = miestai_url[0]
miestai
```

```python
miestai.set_index('Miestas', inplace=True)
```

```python
miestai1 = miestai.drop('Eilė', axis=1, inplace=True)
```

```python
miestai.info()
```

```python
col_list = miestai.columns.tolist()
```

```python
col_list
```

```python
# Kad būtų patogiau, persivadinkim stulpelius
new_list = []
for item in col_list:
    fixed = item.replace('\xa0m.', '')
    new_list.append(fixed)

miestai.columns = new_list
```

```python
miestai
```

```python
#  matyti, kad kai kuriu metu reikšmės yra string, pasitikrinkime, ar bent viena yra int:
any([isinstance(x, int) for x in miestai['1970']])
```

```python
# naudojame list comprehension, patikrinti kiek reikšmių yra ne integer
len([not isinstance(x, int) for x in miestai['1897']])
```

```python
#bandome ištaisyti reikšmes, kurios leidžiasi ištaisomos: 
def fix_years(x):
    try:
        return int(x)
    except:
        return x
```

```python
miestai['1970'] = miestai['1970'].apply(fix_years)
```

```python
miestai['1897'] = miestai['1897'].apply(fix_years)
```
