# Trūkstami duomenys

```python
df = pd.DataFrame(np.random.rand(7,7), 'a,b,c,d,e,f,g'.split(','), 'S,U,V,W,X,Y,Z'.split(','))
df

# df = pd.DataFrame(np.random.rand(5,6), 
#                   ['a', 'b', 'c', 'd', 'e'], 
#                   ['U', 'V', 'W', 'X', 'Y', 'Z'])

```

valymas

```python
purvini = df[df>0.2]
purvini
```

Funkcija, padėsianti mums aptikti nulines reikšmes yra .isnull(). Naudojimo pavyzdžiai:

```python
purvini.isna()
```

```python
# grąžina lentelę, kur true reikšmės yra NaN
purvini.isnull()
```

```python
purvini.isnull().sum()
```

```python
# Rodo kiek kuriame stulpelyje NaN reikšmių
purvini['Y'].isnull().sum()
```

```python
# Galima tikrinti atskiruose stulpeliuose
purvini['Y'].isnull()
```

pirmas metodas tvarkytis su trūkstamais duomenimis yra .dropna(). Jis išmeta visas eilutes, kuriose yra bent viena NaN reikšmė:

```python
# ismeta visas eilutes. kuriose yra nan - not a number
purvini.dropna()
```

```python
# Galime išmesti tik tas eilutes, ar stulpelius, kurie turi mažiau, negu (pvz) 5 NE NaN reikšmes:
# jeigu yra bent 5 sveiki duomenys, nemetam
purvini.dropna(thresh=5)
```

```python
# jeigu norime išmesti stulpelius, turinčius bent vieną NaN, turime nurodyti ašį.
# jeigu yra bent 5 sveiki duomenys stulpelyje, nemetam
purvini.dropna(thresh=6, axis=1)
```

Papildomai nurodžius ašį, tą patį galima padaryti ir su stulpeliais.
Jei norime NaN reikšmes pakeisti kažkuo kitu, naudosime .fillna() metodą:


```python
purvini.fillna('kiauras')
```

Daugiau prasmės būtų pakeisti pvz. stulpelio vidurkiu:

```python
print(purvini['S'])
purvini['S'].fillna(value=purvini['S'].mean())
# i ta vieta, kur buvo zodis "kiauras" irase 0.570320, t.y. vidurki viso stulpelio
# pakeiciam neegzistuojanti duomeni stulpelio vidurkiu, tik stulpelyje S
```

```python
# uzpildom visas NaN reiksmes ju stulpeliu vidurkiais
purvini.fillna(purvini.mean())
```

# Grupavimas

```python
duomenys = {
    'Salis': str('Lietuva,'*3+'Latvija,'*3+'Estija,'*3)[:-1].split(','),
    'Miestai': 'Vilnius,Kaunas,Klaipeda,Riga,Venspilis,Daugpils,Tallinn,Tartu,Parnu'.split(','),
    'Gyventojai': [541, 287, 147, 716, 43, 105, 400, 101, 46]
}
miestai = pd.DataFrame(duomenys)
miestai
```

panaudokime .groupby() metodą, viduje nurodydami stulpelį, pagal kurį grupuosime.

```python
miestai.groupby('Salis')
```

gavome groupby objektą, kuris turi savo metodus. Tam kad galėtumėm jais naudotis, priskirkime objektą kintamąjam:

```python
countries = miestai.groupby('Salis')
countries.mean(numeric_only=True)
```

```python
# isspausdina sugrupuota zodyna
countries.groups
```

```python
# isspausdina raktazodziu sarasa
countries.groups.keys()
```

```python
# .mean() skaičiuos mums vidurkį:
countries.mean(numeric_only=True)
```

```python
# .sum() sumuos:
countries.sum(numeric_only=True)
```

```python
# iš groupby objekto galime traukti reikšmes tokiu būdu:
# .count() suskaičiuos kiek yra įrašų grupės stulpeliuose:
countries.count()

# atkreipkite dėmesį, kad jeigu metodas negali atlikti veiksmo su stulpelio įrašais, rezultate to stulpelio ir nerodo. 
# šiuo atveju galime suskaičiuoti string įrašus, todėl stulpelis 'Miestas' įtrauktas į rezultatą
```

```python
countries.min()
```

```python
countries.max()
```

šiuo atveju matome, kad miestų stulpelį išrūšiavo pagal abėcelę, o skaičius, kaip ir tikėjomės. Max ir min nėra labai praktiški su string tipo reikšmėmis.

```python
# iš groupby objektų galime traukti reikšmes.
# konkrecios grupes duomenu traukimas
countries.sum(numeric_only=True).loc['Lietuva']
```

.describe() metodas duoda pagrindinę informaciją apie lentelės duomenis:

```python
countries.describe()
# std - standart deviation - vidutinis kvadratinis nuokrypis
```

galime sukeisti stulpelius su eilutėmis su .transpose():

```python
countries.describe().transpose()
```

jeigu domina vieno kurio nors stulpelio statistika, galime ją ištraukti taip

```python
countries.describe().transpose()['Estija']
```

# DF Jungimas

```python
mtop = miestai[:6]
mtop
```

```python
mbot = miestai[6:]
mbot
```

norėdami sujungti data_main su data_bottom naudosime .concat() metodą, viduje įrašydami lentelių, kurias jungsime sąrašą:

```python
pd.concat([mtop, mbot])
```

galime prijungti lentelę iš dešinės:

```python
pd.concat([mtop, mbot], axis=1)
```

matome, kad pandas ieškojo antroje lentelėje sutampančių indeksų, ir neradus atliko jungimą užpildant tuščias vietas NaN. Todėl jungiant, reikia įsitikinti, kad jungimui tinka abi pusės. Pertvarkykime data_right indeksą:

```python
mrigth = mbot.reset_index()
mrigth.drop(columns='index', inplace=True)
mrigth
```

```python
pd.concat([mtop, mrigth], axis=1)
```

```python
lenkija = {
    "Salis": ",".join(["Lenkija" for x in range(3)]).split(','),
    "Miestai": "Varsuva,Vroclavas,Gdanskas".split(','),
    "Gyventojai": [1688, 638, 461]
}
polonia = pd.DataFrame(lenkija)
polonia
```

```python
mpl = pd.concat([mrigth, polonia]).reset_index().drop(columns='index')
mpl
```

Pabandykime sujungti data_main su data_right :)

```python
pd.concat([mtop, mpl], axis=1)
```

.merge()
Jeigu data_main ir data_right turėtų po vieną sutampantį stulpelį, pvz.:

```python
key = 'A,B,C,D,E,F'.split(',')
mtop['key'] = key
mpl['key'] = key
```

```python
pd.merge(mtop, mpl, on='key')
```

matome, kad lentelės suklijuotos, tačiau stulpelis 'key' nesikartoja (stulpeliai pervadinti, kad nesidubliuotų pavadinimai). Stulpelį 'key' teko nurodyti parametre on.

papildykime pavyzdį dar vienu stulpeliu:

```python
key2 = 'B,C,D,E,F,G'.split(',')
mtop['key2'] = key2
mpl['key2'] = key2
```

```python
mtop, mpl
```

```python
pd.merge(mtop, mpl, on=['key', 'key2'])
```

Matome, kad į parametrą on galima perduoti ir daugiau reikšmių, sąraše.
Jeigu, pvz. stulpelis 'key2' vienoje lentelėje skiriasi nuo kitos lentelės 'key2' stulpelio:

```python
keyx = 'W,C,D,E,F,W'.split(',')
mpl['key2'] = keyx
mtop, mpl
```

...gauname tik tas eilutes, kuriose 'key2' reikšmių sekos dalis sutampa.
Yra ir metodas .join(). Jis daro tą patį, ką ir merge, tik indekso pagrindu. Pvz.:

```python
mtop.join(mpl, lsuffix='L', rsuffix='R')
```

Kadangi sutampa stulpelių pavadinimai, teko nurodyti lsuffix ir rsuffix. Jeigu indeksas nebūtų identiškas, rezultate gautumėm tik tas eilutes, kurių indeksas persidengia. Dažniausiai vis tik naudosime .merge()
