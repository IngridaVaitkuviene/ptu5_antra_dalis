# Tiesinė regresija užduotis

Turime lentelę su 7 žuvų rūšim, bei įvairiais jų išmatavimais. Vienas iš jų yra svoris - tai bus mūsų target stulpelis. Sukurkite tiesinės regresijos modelį, kuris pagal kitus lentelės stulpelius spėlios žuvies svorį. Išsiaiškinkite, ar yra trūkstamų reikšmių, atsispausdinkite keletą diagramų ir histogramų, patyrinėkite tendencijas ir koreliacijas.

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
fish = pd.read_csv('https://raw.githubusercontent.com/robotautas/kursas/master/Machine%20Learning/Fish.csv')
fish.head()
```

```python
fish.info()
```

```python
dummies = pd.get_dummies(fish['Species'])
dummies.head()
```

```python
data = pd.concat([fish, dummies], axis=1)
data.tail()
```

```python
data.drop(['Species'], axis=1, inplace=True)
data.tail()
```

```python
features = data[list(data.columns)[1:]]
features
```

```python
target = data['Weight']
target
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.60, random_state=42
)
features_train
```

```python
target_train[69]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ai = LinearRegression().fit(features_train, target_train)
```

```python
predictions = ai.predict(features_test)
predictions
```

```python
data_predictions = pd.Series(data=predictions, name='fish_predictions')
data_predictions
```

```python
comparison = pd.concat([target_test.reset_index(), data_predictions], axis=1)
comparison
```

```python
ai.score(features_test, target_test)
```

```python
final_data = pd.merge(fish[['Species', 'Weight']].reset_index(), comparison, on='index')
final_data
```
